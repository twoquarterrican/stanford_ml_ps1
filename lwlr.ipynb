#### Weight function

The "locally-weighted" in locally-weighted linear regression comes from using a weight function which depends on $x$.  In general, any "bump" function centered on x will work.  A "bump" function centered at $x$ is a positive function which decreases with distance from $x$.  A Gaussian is an example of a bump function.  The weight/bump function below is Gaussian-type.  This means it is not a probability distribution, i.e. does not integrate to 1.  Some consider this a necessary requirement for being called a Gaussian.  It is a bump function which is 1 at $x$ and quickly falls away toward 0 at points far from $x$.  The parameter $tau$ controls the quickness with which the weight $w(x)$ goes to 0.  If $tau$ is large, $w$ falls off slowly.  If $\tau$ is small, $w$ falls off quickly.  Thus a smaller value of $\tau$ gives a more local predictive model.  At the end of this notebook we will observe several values of $\tau$ and see that the predictions become more uniform as $\tau$ gets bigger.

```python
# We weigh the sample points close to x heavier than those far from x
def compute_weights(X, x, tau):
    "Assume X is (k by N), x is (1 by N), and tau is a non-zero real number."
    "This function returns a (1 by k) matrix whose i^{th} entry is exp(-|x-x(i)|^2/(2tau^2)) \
    where x(i) is the i^{th} row of X"
    
    # the difference of a (k by N) matrix X and a (1 by N) matrix x is the (k by N) matrix
    # whose i^th row and j^th column is X[i,j]-x[j]
    diff = X - x
    
    # diff*diff.T is (k by k) and has diagonal entries given by the norms of the row vectors of diff, 
    # output of .diagonal() is a row matrix.  In this case, output is (1 by k)
    # np.exp() exponentiates each entry individually when applied to non-square matrices
    return np.exp(-(diff*diff.T).diagonal()/(2*(tau**2)))
```

Confirm that the weight function is behaving as it should. We plot the sample points, adjusting down the opacity of the points with lower weight.  We should see a fuzzy picture which is centered around a certain point.  We test on the test points from above.

```python
# function to plot the training set with opacity dictated by local weight.  Local is with respect to argument x
def plot_points_with_local_weight(x, tau):
    # compute_weights returns a 1xm matrix, so get a size m array from it
    grey_by_weight = np.asarray(compute_weights(x_sample_ones, x, tau))[0]
    # loop through to define each opacity (alpha)
    for i in range(m):
        plt.scatter(x_sample[i,0],x_sample[i,1], alpha=grey_by_weight[i], c=ycol[i], cmap=cm.Greys, marker='o')
    # plot, using a red dot, the point around which the weights are centered.
    plt.plot(x[0,1],x[0,2],marker='o', color='r')
    # normalize axes
    plt.axis([-1.01,1.01,-1.01,1.01])
```

### The Newton-Raphson Algorithm

#### This algorithm should be thought of as a function $\theta_i \mapsto \theta_{i+1}$, for fixed training set X_train, y_train, point x, and parameter tau

```python
def compute_h_theta_x(X, theta):
    "input: X is (k by N), theta is (1 by N)"
    "output: (k by 1) matrix whose i^th entry is the logistic function applied to the dot product of theta with the i^th row of Xb"
    return 1/(1+np.exp(- X*theta.T))
    
# test that we get the right size matrix and test the first entry
X=x_test
theta = np.matrix([[0.5, 1.2, 3.8]])
print "first entry of result should be " + str(1/(1+np.exp( - X[0,0]*theta[0,0] - X[0,1]*theta[0,1] - X[0,2]*theta[0,2])))
compute_h_theta_x(X,theta)
```

We need to deform locally-weighted linear regression action by a parameter lambda to make the Newton-Raphson algorithm converge.

```python
lambda_const = 0.0001
```

The Newton-Raphson algorithm for optimizing $\ell(\theta)$:  Pick $\theta_0$, then inductively set

$$\theta_i = \theta_{i-1} - H^{-1} \nabla_{\theta}\ell(\theta),$$

then $\lim_{i\to \infty} \theta_i$ is our optimal solution.  Here, $H$ is the Hessian of $\ell$ and $\nabla_\theta \ell$ is the gradient.  Below, our matrices are the transposes of what we have in this description, so $H^{-1}$ is on the right side of the gradient in the code given.  The comments give dimensions of matrices to make sense of this.

```python
def newton_raphson(theta,X_train,y_train,x,tau):
    # dimensions: theta (1 by N), X_train (m by N), y_train (m by 1), x (1 by N)
    # tau is a non-zero real number
    
    # we need the gradient of \ell and the Hessian.
    # the gradient is -\lambda*\theta_0 + X_train.T*z, 
    # where z = W*y_minus_h, where W is the Weight matrix, which does not depend on theta.
    # y_minus_h is below
    
    # W is (1 by m), an instance of class np.matrix
    W = compute_weights(X_train, x, tau)
    
    # given theta, the important values are h(theta, x(i)) for each i, we make a diagonal (m by m) matrix Diag_y_minus_h with these values
    # compute... returns (m by 1) matrix, and np.diagflat converts this into a diagonal (m by m) matrix
    Diag_h_theta_X_train = np.diagflat(compute_h_theta_x(X_train, theta))
                                       
    # the column vector y_minus_h has i^{th} row given by y(i) - h(theta,x(i)). This is (m by 1), but we think of it as
    # diagonal (m by m) so that we can use matrix multiplication to scale each y(i) - h(theta,x(i)) by the weight w(i)(x) 
    Diag_y_minus_h = np.diagflat(y_train) - Diag_h_theta_X_train
    
    # z is (1 by m), obtained from (1xm)*(mxm)
    # z_i = w(i)(x)(y(i)-h(theta,xi)), that is, z = W*Diag_y_minus_h
    z = W*Diag_y_minus_h
    
    # now we can compute the gradient, which is -lambda*theta + z*X_train
    # X_train is (m by N), z is (1 by m), so z*X_train is (1 by N).  Note that this is the assumed dimension of theta.
    # lambda is a real numbe. grad is (1 by N).
    grad = -lambda_const*theta + z*X_train
    
    # the Hessian (N by N) is -\lambda*I_N + X_train.T * D * X_train, where D (m by m) is diagonal with ith entry given by h(theta, x(i))(1-h(theta,x(i)).
    # I_N is the N by N identity matrix.  Note that the product X_train.T * D * X_train has dimension (Nxm)(mxm)(mxN) = (NxN)
    
    # build the diagonal matrix, np.eye(k) is identity matrix of size k
    D = Diag_h_theta_X_train*(np.eye(m) - Diag_h_theta_X_train)
    
    # build the Hessian
    Hess = -lambda_const*np.eye(N) + X_train.T*D*X_train
    # invert the Hessian
    Hess_inv = Hess.getI()
    # update theta: dimensions are (1xN) - (1xN)(NxN)
    theta = theta - grad*Hess_inv
    
    # note that the solutions x of the equation theta.T*x are independent of the length of theta.  If we want theta to converge, we should normalize it.
    # If theta = 0 we get an error, but in that case something has gone wrong with the algorithm, so we don't test for that problem
    return theta/np.linalg.norm(theta)
```

A utility function to plot the decision boundary, which by definition is the set of $x \in \mathbb R^{N-1}$ such that $\theta^{T} (1,x) = 0$.

```python
def plot_theta_line(theta):
    # theta is 1x3, convert it to size 3 array
    theta = np.asarray(theta)[0]
    # plot line theta[0] + theta[1]x + theta[2]y = 0
    if(theta[2]!=0):
        y1 = (theta[0]+theta[1])/theta[2]
        y2 = (theta[0]-theta[1])/theta[2]
        plt.plot([-1,1],[y1, y2])
    elif(theta[1]!=0):
        x1 = -theta[0]/theta[1]
        plt.plot([x1,x1],[-1,1])
```

Test one iteration of the Newton-Raphson algorithm with initial guess $\theta = 0$.  Look at each of the sample points from above. Testing more iterations of the Newton-Raphson algorithm.  Define a function which takes a fixed number of iterations, $n$, as an argument.

```python
def test_iterate_newton_raphson(x,tau,n):
    # n is number of iterations
    plot_points_with_local_weight(x,tau)
    theta = theta_test
    for i in range(n):
        theta = newton_raphson(theta, X_train, y_train, x, tau)
        plot_theta_line(theta)
```

Before we chose a fixed number of iterations.  Let us choose a tolerance parameter $\epsilon$ and run the algorithm until $\theta_{i} - \theta_{i-1}$ has norm smaller than $\epsilon$.  We also insert a safety so we don't get stuck in an infinite loop in the case the code has a mistake.

```python
def test_converge_newton_raphson(x, tau, tolerance):
    # loop until next iteration is within tolerance of previous
    plot_points_with_local_weight(x, tau)
    theta0 = theta_test
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 500):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        plot_theta_line(theta1)
        theta0 = theta1        
```

Iterate with moderate value of $\tau$ until convergence up to given tolerance.

```python
run_procedure_on_test_points(test_converge_newton_raphson, 0.5, 0.001)
```

Categorize each of the test points with respect to a given tolerance and locality parameter.  This is roughly the same as what was done above, except that we only draw the final decision boundary, not all of the iterates.

```python
def test_find_decision_boundary_newton_raphson(x, tau, tolerance):
    # loop until next iteration is within tolerance of previous
    plot_points_with_local_weight(x, tau)
    theta0 = theta_test
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 500):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        theta0 = theta1
    plot_theta_line(theta1)
```

The graphs below do not show explicitly which side of the decision boundary corresponds to predicting the value 1 for the query point, but it should be clear from the picture.

```python
run_procedure_on_test_points(test_find_decision_boundary_newton_raphson, 0.5, 0.001)
```

The same code as above, but without plotting the decision boundary.

```python
def compute_theta(X_train, y_train, x, tau, tolerance):
    theta0 = np.matrix([[0,0,0]])
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 50):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        theta0 = theta1
    return theta1
```

Turn a given $\theta$ into a decision: given $x$, should $y$ be $0$ or $1$?

```python
def classify_point(theta,x):
    prediction = compute_h_theta_X_train(theta,x)
    if(prediction > 0.5):
        return 1
    else:
        return 0
```

The final classification.  The upshot is that if we had used linear regression instead of locally-weighted linear regression, we would have a straight line dividing the two categories.  The local weighting allows more interesting predictions.

```python
def categorize_all(tau):
    X_train = np.asmatrix(x_sample_ones)
    y_train = np.asmatrix(y_sample).T
    tolerance = 0.001
    samples_per_x = 80
    samples_per_y = 65
    xv, yv = np.meshgrid(linspace(-1,1,samples_per_x),linspace(-1,1,samples_per_y))
    square_colors = np.zeros(shape=(samples_per_y,samples_per_x))+50
    for i in range(samples_per_y):
        for j in range(samples_per_x):
            x = np.matrix([[1, xv[i,j], yv[i,j]]])
            square_colors[i,j] = classify_point(compute_theta(X_train, y_train, x, tau, tolerance), x)
    plt.axis([-1.01,1.01,-1.01,1.01])
    plt.scatter(xv,yv,marker='s',edgecolors='none', c=square_colors)
```

A reminder of the data, followed by a classification of a grid of points with varying values of the localization parameter $\tau$.  Clearly large $\tau$ means a more uniform prediction and small $\tau$ means a prediction more sensitive to local data.
