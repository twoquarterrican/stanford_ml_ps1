#### Weight function

```python
# We weigh the sample points close to x heavier than those far from x
def compute_weights(X_train, x, tau):
    """ if row i of X_train is x(i), then x(i) and x both live in R^N, viewed as row vectors.  This function returns a 1xm matrix whose i^{th} entry is exp(-|x-x(i)|^2/(2tau^2))"""
    # assume first column of X_train is all 1's, same for x
    # assume x is 1xN, and X_train is mxN
    
    # the difference of an mxN matrix and a 1xN matrix is an mx1 matrix
    # whose i^th row and j^th column is X_train[i,j]-x[j]
    diff = X_train - x
    
    # diff*diff.T has diagonal entries given by the norms of the row vectors
    # np.exp() exponentiates each entry individually when applied to non-square matrices
    return np.exp(-(diff*diff.T).diagonal()/(2*(tau**2)))

def weight(xi, x, tau):
    if((xi.shape != x.shape) or (x.shape[1] != 1)):
       raise Exception("first two arguments of weight must be column vectors of the same size")
    # input: tau real, xi, x in R^N
    # output w(i)(x) in R
    delta = x - xi
    return exp( - delta.T.dot(delta) / (2*tau**2) )[0,0]
```

### The Newton-Raphson Algorithm

#### This algorithm should be thought of as a function $\theta_i \mapsto \theta_{i+1}$, for fixed training set X_train, y_train, point x, and parameter tau

```python
def compute_h_theta_X_train(X_train, theta):
    """input: X_train is m x N matrix, theta is 1 x N matrix"""
    """output: an m x 1 matrix whose i^th entry is the logistic function applied to the dot product of the i^th row of X_train and theta"""
    return 1/(1+np.exp(- X_train*theta.T))
    
theta_test = np.matrix([[0.5, 1.2, 3.8]])
np.diagflat(compute_h_theta_X_train(x_test, theta_test))
```

```python
# We need to deform locally-weighted linear regression action by a parameter lambda to make the Newton-Raphson algorithm converge
lambda_const = 0.0001
```

```python
def newton_raphson(theta,X_train,y_train,x,tau):
    # dimensions: theta is 1xN, X_train is mxN, y_train is mx1, x is 1xN, tau is number
    # we need the gradient of \ell and the Hessian
    # the gradient is -\lambda*\theta_0 + X_train.T*z, 
    # where z = W*y_minus_h, where W is the Weight matrix, which does not depend on theta, theta is 1xN.
    # y_minus_h is below
    
    # W is 1 by m, of class np.matrix
    W = compute_weights(X_train, x, tau)
    
    # given theta, the important values are h(theta, x(i)) for each i, we make a diagonal mxm matrix D_y_minus_h with these values
    Diag_h_theta_X_train = np.diagflat(compute_h_theta_X_train(X_train, theta))
                                       
    # the column vector y_minus_h whose ith row is y(i) - h(theta,x(i)), y_minus_h is an mx1 Matrix, but we think of it as
    # an mxm diagonal matrix, so we can scale each y(i) - h(theta,x(i)) by the weight w(i)(x) using matrix multiplication
    Diag_y_minus_h = np.diagflat(y_train) - Diag_h_theta_X_train
    
    # z is 1xm, obtained from (1xm)*(mxm) matrix multiplication.
    # z_i = w(i)(x)(y(i)-h(theta,xi)), or z = W*Diag_y_minus_h
    z = W*Diag_y_minus_h
    
    # now we can compute the gradient, which is -lambda_ * theta + z*X_train
    # X_train is mxN and z is 1xm, so z*X_train is 1xN, this is the assumed dimension of theta
    # dim grad = 1xN
    grad = -lambda_const*theta + z*X_train
    
    # the Hessian is -\lambda I_N + X_train.T * D * X_train, where D is mxm diagonal matrix with ith entry given by h(theta, x(i))(1-h(theta,x(i)), and
    # I_N is the NxN identity matrix.  note that the product X_train.T * D * X_train has dimension (Nxm)(mxm)(mxN) = (NxN)
    
    # build the diagonal matrix
    D = Diag_h_theta_X_train*(np.eye(m) - Diag_h_theta_X_train)
    
    # vuild the Hessian
    Hess = -lambda_const*np.eye(N) + X_train.T*D*X_train
    # invert the Hessian
    Hess_inv = Hess.getI()
    # update theta: dimensions are (1xN) - (1xN)(NxN)
    theta = theta - grad*Hess_inv
    
    return theta/np.linalg.norm(theta)
```

```python
def plot_theta_line(theta):
    # theta is 1x3, convert it to size 3 array
    theta = np.asarray(theta)[0]
    # plot line theta[0] + theta[1]x + theta[2]y = 0
    if(theta[2]!=0):
        y1 = (theta[0]+theta[1])/theta[2]
        y2 = (theta[0]-theta[1])/theta[2]
        plt.plot([-1,1],[y1, y2])
    elif(theta[1]!=0):
        x1 = -theta[0]/theta[1]
        plt.plot([x1,x1],[-1,1])
```

```python
theta_test = np.matrix([[0,0,0]])
X_train = np.asmatrix(x_sample_ones)
y_train = np.asmatrix(y_sample).T
def test_newton_raphson(x,tau):
    plot_points_with_local_weight(x,tau)
    plot_theta_line(newton_raphson(theta_test, X_train, y_train, x, tau))
```

```python
# note that after 1 iteration of the algorithm starting from theta = 0, we get different separating lines for each test point
run_procedure_on_test_points(test_newton_raphson, 0.5)
```

```python
def test_iterate_newton_raphson(x,tau,n):
    # n is number of iterations
    plot_points_with_local_weight(x,tau)
    theta = theta_test
    for i in range(n):
        theta = newton_raphson(theta, X_train, y_train, x, tau)
        plot_theta_line(theta)
```

```python
# this gives an idea on how the local decision boundaries change as we iterate
run_procedure_on_test_points(test_iterate_newton_raphson, 0.5, 9)
```

```python
# Higher tau means the decision boundaries are more similar
run_procedure_on_test_points(test_iterate_newton_raphson, 1, 9)
```

```python
def test_converge_newton_raphson(x, tau, tolerance):
    # loop until next iteration is within tolerance of previous
    plot_points_with_local_weight(x, tau)
    theta0 = theta_test
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 50):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        plot_theta_line(theta1)
        theta0 = theta1        
```

```python
# iterate until convergence up to given tolerance
run_procedure_on_test_points(test_converge_newton_raphson, 0.5, 0.001)
```

```python
def test_find_decision_boundary_newton_raphson(x, tau, tolerance):
    # loop until next iteration is within tolerance of previous
    plot_points_with_local_weight(x, tau)
    theta0 = theta_test
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 50):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        theta0 = theta1
    plot_theta_line(theta1)
```

```python
# just show the decision boundaries
run_procedure_on_test_points(test_find_decision_boundary_newton_raphson, 0.5, 0.001)
```

```python
def compute_theta(X_train, y_train, x, tau, tolerance):
    theta0 = np.matrix([[0,0,0]])
    safety = 0
    while(True):
        safety = safety + 1
        if (safety > 50):
            break
        theta1 = newton_raphson(theta0, X_train, y_train, x, tau)
        # each thetai is 1x3
        difference = theta1 - theta0
        # note: it is important that theta1 and theta2 have the same norm, otherwise this is a bad convergence test
        if(numpy.linalg.norm(difference) < tolerance):
            break
        theta0 = theta1
    return theta1
```

```python
def classify_point(theta,x):
    prediction = compute_h_theta_X_train(theta,x)
    if(prediction > 0.5):
        return 1
    else:
        return 0
```

```python
def color(number):
    if(number == 1):
        return 'red'
    elif(number == 0):
        return 'blue'
    else:
        return 'black'
```

```python
# the final classification.  The upshot is that if we had used linear regression
# instead of locally-weighted linear regression, we would have a straight line 
# dividing the two categories.  The local weighting allows the line to curve
def categorize_all(tau):
    X_train = np.asmatrix(x_sample_ones)
    y_train = np.asmatrix(y_sample).T
    tolerance = 0.001
    samples_per_x = 80
    samples_per_y = 65
    xv, yv = np.meshgrid(linspace(-1,1,samples_per_x),linspace(-1,1,samples_per_y))
    square_colors = np.zeros(shape=(samples_per_y,samples_per_x))+50
    for i in range(samples_per_y):
        for j in range(samples_per_x):
            x = np.matrix([[1, xv[i,j], yv[i,j]]])
            square_colors[i,j] = classify_point(compute_theta(X_train, y_train, x, tau, tolerance), x)
    plt.axis([-1.01,1.01,-1.01,1.01])
    plt.scatter(xv,yv,marker='s',edgecolors='none', c=square_colors)
```
